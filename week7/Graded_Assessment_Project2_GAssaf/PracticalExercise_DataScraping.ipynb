```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
pages = set()


url="https://www.canada.ca/en/public-health/services/surveillance/respiratory-virus-detections-canada/2018-2019/respiratory-virus-detections-isolations-week-01-ending-january-5-2019.html"
html = urlopen(url)
bs = BeautifulSoup(html, 'html.parser')

# selecting the table with id=2
table = bs.find('table', {'id': 't2'})
table
```

```python
df_headers=[] # retrieve the header from the table since they are under the tag <th>
table_headers=table.find_all('th')
for header in table_headers:
    df_headers.append(header.get_text(strip=True))
df_headers
```

```python
import pandas as pd
data=[]
for row in table.find('tbody').find_all('tr')[1:]:  #table.find('tbody') to select the rows under  the table body only. Skip header row and looping through all the table rows <tr>
    columns = row.find_all('td') # retrieving all the columns under the row
    if columns:  # Only add rows with data
        row_data = [col.get_text(strip=True) for col in columns] # get the data from different columns
        data.append(row_data) #add the data available under columns to data

# Create a DataFrame
df = pd.DataFrame(data, columns=df_headers)

# Display the DataFrame
df.head(100)
```

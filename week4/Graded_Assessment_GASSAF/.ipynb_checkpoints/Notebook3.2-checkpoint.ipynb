Question 1: Identify all the outliers in the above data.  Justify your answers using objective measures.

```python
#Knowing that the salary column can be considered as a number column. removing the $ and k sign is mandatory before transforming this to int
grads['Salary'] = grads['Salary'].str.replace('$', '').str.replace('k', '').astype(int)
```

```python
numerical_columns = ["Avg Hours Studies per Week", "GPA", "Sense of Humour (0-5)","Salary"]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))

for i, col in enumerate(numerical_columns, start=1):
    plt.subplot(2, 2, i)
    plt.boxplot(grads[col].dropna(), vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
    plt.title(col)
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

From the above boxplot, there is only an outlier available under the column Avg Hours Studies per Week

Another method is to find the outliers using IQR method.

```python
def find_outliers_iqr(column,col):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    print(f"For column {col}:\n    lower_bound is {lower_bound} \n    upper_bound is: {upper_bound}")
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return outliers

# Identify outliers for numerical columns
outliers = {}
numerical_columns = ["Avg Hours Studies per Week", "GPA", "Sense of Humour (0-5)","Salary"]
for col in numerical_columns:
    outliers[col] = find_outliers_iqr(grads[col],col)

# Print outliers
for col, outlier_values in outliers.items():
    print(f"Outliers in {col}:")
    print(outlier_values)
    print()
```

Question 3: Reload the data and fill-in the data using mean method as well as the frequent method.

```python
grads = pd.read_csv("https://raw.githubusercontent.com/harmanani/AAI614/main/Week%203/grads.csv")
```

Applying the mean method:

```python
from sklearn.impute import SimpleImputer

# Initialize the SimpleImputer with strategy='mean'
imputer = SimpleImputer(strategy='mean')

# Apply the imputer to numeric columns with missing values
numeric_columns_with_missing_values = ['GPA', 'Sense of Humour (0-5)']
grads[numeric_columns_with_missing_values] = imputer.fit_transform(grads[numeric_columns_with_missing_values])
```

the missing value in GPA is fill with the mean of the column that equal to 3.11 
and the 2 missing values in sense of humour are filled in with the mean of column that 
equal to 2.27

Applying the frequent method:

```python
grads = pd.read_csv("https://raw.githubusercontent.com/harmanani/AAI614/main/Week%203/grads.csv")
```

```python
from sklearn.impute import SimpleImputer

# Initialize the SimpleImputer with strategy='most_frequent'
imputer = SimpleImputer(strategy='most_frequent')

# Apply the imputer to numeric columns with missing values
numeric_columns_with_missing_values = ['GPA', 'Sense of Humour (0-5)']
grads[numeric_columns_with_missing_values] = imputer.fit_transform(grads[numeric_columns_with_missing_values])
```

Exercise II. Run the cell below to create a new dataframe called `df_miss`.  Its first column will contain some missing values.

```python
import pandas as pd
import numpy as np
import random

nrows = 10
ncols = 5

# set a seed for random number generation
np.random.seed(314)
# create an array filled with random data
data = np.array(np.random.rand(nrows, ncols))
# put the data to a pandas dataframe
df_miss = pd.DataFrame(data)
# rename the columns
df_miss.columns = ['col_'+str(ii) for ii in range(ncols)]

# randomly set some values to missing
ix0 = np.random.randint(nrows, size=3)
ix1 = np.random.randint(nrows, size=3)

df_miss['col_0'][ix0] = np.nan
df_miss['col_1'][ix1] = np.nan

print(df_miss)
```

Impute the missing values (NaN) in `col_0` (but not `col_1`) with the median.  Store the values in the dataframe by using the parameter `inplace`.  Print the dataframe.

```python
# Fill missing values in 'col_1' with the median value of that column
df_miss['col_0'].fillna(df_miss['col_0'].median(), inplace=True)
df_miss
```

Impute the missing values in `col_1` with value 0.  Store the values in the dataframe by using the parameter `inplace`.  Print the dataframe.

```python
df_miss['col_1'].fillna(0, inplace=True)
df_miss
```

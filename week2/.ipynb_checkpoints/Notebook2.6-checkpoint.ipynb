```python
df1=pd.read_csv(filenames[0]) ## Added by Georges Assaf to read the columns name available in the first csv
df1
```

```python
columns_use=df1.columns.tolist()
all_dfs = [pd.read_csv(one_filename,header=None,names=columns_use, usecols=range(18),skiprows=1) for one_filename in filenames[0:]]  # reading with skipping the first row
```

```python
dff=pd.concat(all_dfs,ignore_index=True)
dff
```

```python
df=dff
```

```python
df_backup=df
```

```python
# Create a new column, pre_tip_amount
df['pre_tip_amount'] = (
    df
    [['fare_amount', 'extra', 'mta_tax', 'tolls_amount', 
      'improvement_surcharge', 'congestion_surcharge']]
    .sum(axis='columns')
)
```

```python
list(df.columns)
```

```python
# Create a new column, tip_percentage, showing the percent of the fare_amount that people tipped
df['tip_percentage'] = df['tip_amount'] / df['pre_tip_amount']
```

```python
# What was the overall tip percentage?
df['tip_percentage'].mean()
```

```python
list(df.columns)
#df['tip_percentage']
```

```python
# How many times did people tip more than 100%?
(df['tip_percentage'] > 1).value_counts()
print(f"people tip  more than 100%,   {(df['tip_percentage'] > 1).value_counts()[1]} times")
```

```python
# What percent of taxi riders give no tip at all?
(df['tip_percentage'] == 0).value_counts(normalize=True)
print(f"the percentage of taxi riers that give no tip at all is {(df['tip_percentage'] == 0).value_counts(normalize=True)[1]*100:.2f}%")
```

```python
df.dtypes
```

# Steps added by ME [Georges ASSAF] to convert the column tpep_pickup_datetime to datetime type

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'], errors='coerce') # Added to conver the column fro object to datatime format
```

```python
list(df.columns)
```

```python
# On which day of the week do people tip the greatest percentage?
# Mon = 0, Tues = 1, Wed = 2, Thu = 3, Fri = 4, Sat = 5, Sun = 6

df.groupby(df['tpep_pickup_datetime'].dt.day_of_week)['tip_percentage'].mean().sort_values(ascending=False)
```

```python
import datetime as dt
# At which hour do people tip the greatest percentage?
df.groupby(df['tpep_pickup_datetime'].dt.hour)['tip_percentage'].mean().sort_values(ascending=False)
```

```python
# Do people tip more, on average, in January or July?
df.groupby(df['tpep_pickup_datetime'].dt.month)['tip_percentage'].mean().sort_values(ascending=False)
```

# Questions

1. We saw that 32 percent of riders don't tip at all. Of those who *do*, what percentage do they tip, on average?

```python
# Filter out rows where tip_percentage is 0
df_tipped = df[df['tip_percentage'] > 0]

# Calculate the mean tip percentage for riders who tipped
average_tip_percentage = df_tipped['tip_percentage'].mean()

average_tip_percentage
```

```python
average_tip_percentage = df[df['tip_percentage'] > 0]['tip_percentage'].mean()
print(f"Of those who tips, the percentage is {average_tip_percentage*100:.2f} %")
```

2. How many of the rides in our data set, supposedly from January and July 2019, are from outside of those dates?

```python

outside_jan_july = df[~((df.index >= '2019-01-01') & (df.index <= '2019-01-31')) & 
                      ~((df.index >= '2019-07-01') & (df.index <= '2019-07-31'))]
num_outside = outside_jan_july.shape[0]
print(f"the number of rides outside the month of January and July 2019 is {num_outside} rides")
```

3. Looking only at dates in January and July, on what week did passengers tip the greatest percentage?

```python
data_jan_july = df[((df.index >= '2019-01-01') & (df.index <= '2019-01-31')) | 
                      ((df.index >= '2019-07-01') & (df.index <= '2019-07-31'))]
data_jan_july=data_jan_july.reset_index()
data_jan_july['tpep_pickup_datetime'] = pd.to_datetime(data_jan_july['tpep_pickup_datetime'], errors='coerce') # Added to conver the column fro object to datatime format
```

```python
data_jan_july
```

```python
import sys
import subprocess
import pkg_resources

required = {'autoviz'}
installed = {pkg.key for pkg in pkg_resources.working_set}
missing = required - installed

if missing:
    python = sys.executable
    subprocess.check_call([python, '-m', 'pip', 'install', *missing], stdout=subprocess.DEVNULL)
    
import pandas as pd
import numpy as np
from autoviz import AutoViz_Class
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
students = pd.read_csv("https://raw.githubusercontent.com/harmanani/AAI614/main/Week%202/students.csv")
```

##### If you donâ€™t have a target variable you can leave it blank:

```python
target_variable = 'Avg_Hours_Studied_per_week'
```

```python
AV = AutoViz_Class()
sep = ','


dft = AV.AutoViz(students, sep, depVar=target_variable, dfte="",
           header=0, verbose=1,
            lowess=False,chart_format='bokeh',max_rows_analyzed=500,max_cols_analyzed=30)
```

# Converting the salary to numbers after replaving the $ and k with an empty string and then converting it to float

```python
students.head()
# Remove the dollar sign and 'k', then convert to integers
students['Salary'] = students['Salary'].replace('[\$,k]', '', regex=True).astype(float)*1000
students = students.dropna(subset=['GPA'])
students = students.dropna(subset=['Avg_Hours_Studied_per_week'])
students = students.dropna(subset=['Sense of Humour (0-5)'])
students.head()

```

```python
target_variable = 'Avg_Hours_Studied_per_week'
```

```python
AV = AutoViz_Class()
sep = ','


dft = AV.AutoViz(students, sep, depVar=target_variable, dfte="",
           header=0, verbose=1,
            lowess=False,chart_format='bokeh',max_rows_analyzed=500,max_cols_analyzed=30)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

students_numeric = students.select_dtypes(include='number')


correlation_matrix = students_numeric.corr()


#plt.figure(figsize=(10, 8))

# Create a heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)


plt.show()

```
